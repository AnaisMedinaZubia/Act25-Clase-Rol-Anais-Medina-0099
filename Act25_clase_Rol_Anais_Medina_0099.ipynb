```python
class Rol:
    # Constructor: Aquí se definen los 3 atributos
    def __init__(self, nombre, permisos, descripcion):
        self.nombre = nombre
        self.permisos = permisos
        self.descripcion = descripcion

    # Función 1: Leer datos (Permite actualizar la información)
    def leer_datos(self):
        print(f"\n--- Editando datos para el rol: {self.nombre} ---")
        self.nombre = input("Nuevo nombre del rol: ")
        self.permisos = input("Nuevos permisos: ")
        self.descripcion = input("Nueva descripción: ")
        print("¡Datos actualizados con éxito!")

    # Función 2: Mostrar datos (Imprime la información del objeto)
    def mostrar_datos(self):
        print("\n============================")
        print(f"DATOS DEL ROL: {self.nombre}")
        print(f"Permisos: {self.permisos}")
        print(f"Descripción: {self.descripcion}")
        print("============================")

# --- Bloque Interactivo ---

print("Configuración inicial del Rol 1")
rol1 = Rol(
    input("Nombre: "),
    input("Permisos: "),
    input("Descripción: ")
)

print("\nConfiguración inicial del Rol 2")
rol2 = Rol(
    input("Nombre: "),
    input("Permisos: "),
    input("Descripción: ")
)

# Usamos la función Mostrar Datos
rol1.mostrar_datos()
rol2.mostrar_datos()

# Ejemplo de uso de la función Leer Datos para modificar un objeto
respuesta = input("\n¿Deseas modificar los datos del Rol 1? (s/n): ")
if respuesta.lower() == 's':
    rol1.leer_datos()
    rol1.mostrar_datos()

print("\n¡Proceso finalizado!")
print("Programa realizado por Anais medina NL = 36")
```
